```python
import time 
import datetime 

def get_stages(app_id):
    URL = "http://bu-23-101:18080/api/v1/applications/%s/stages"%(app_id)
    # sending get request and saving the response as response object 
    r = requests.get(url = URL) 

    # extracting data in json format 
    data = r.json() 

    print(len(data))
    for stg in data:
        #print(stg)
        #for field in stg:
        #    print(field)
        completion_time = time.mktime(datetime.datetime.strptime(stg['completionTime'], 
                                                     "%Y-%m-%dT%H:%M:%S.%fGMT").timetuple())
        start_time = time.mktime(datetime.datetime.strptime(stg['submissionTime'], 
                                                     "%Y-%m-%dT%H:%M:%S.%fGMT").timetuple())
        
        print(stg['stageId'], stg['inputBytes'], stg['outputBytes'],
                completion_time - start_time)
        #break
        
        
for app in data:
    print(app)
    print(app['id'])
    get_stages(app['id'])
    break
    
    
    
```

```python
colors=['#bdbdbd', '#fdc086', '#d95f02', '#4daf4a']
df1 = pd.read_csv('implementationresults.csv')

df = df1.groupby("order").agg({'Kariz': 'mean', 'MRD': 'mean', 
                               'CP': 'mean', 'infinite': 'mean', 'NoCache': 'mean'})


df = df1.groupby("order").agg({'Kariz': 'mean', 'MRD': 'mean', 
                               'CP': 'mean', 'infinite': 'mean', 'NoCache': 'mean'})



df['KARIZvsMRD'] = df['MRD']/df['Kariz']
df['KARIZvsCP'] = df['CP']/df['Kariz']

df['1Karizp'] = df['NoCache']/df['Kariz']
df['1CPp'] = df['NoCache']/df['CP']
df['1MRDp'] = df['NoCache']/df['MRD']


df3 = df[['1Karizp','KARIZvsMRD','KARIZvsCP']].dropna().reset_index()

print(df3[(df3['order'] == 0) | (df3['order'] == 1)].mean())
print(df3[(df3['order'] == 0) | (df3['order'] == 1)].max(), '\n\n')


print(df3[(df3['order'] >= 2) & (df3['order'] < 5)].mean())
print(df3[(df3['order'] >= 2) & (df3['order'] < 5)].max(), '\n\n')

print(df3[(df3['order'] >= 5) & (df3['order'] < 8)].mean())
print(df3[(df3['order'] >= 5) & (df3['order'] < 8)].max(), '\n\n')

print(df3[df3['order'] == 8].mean())
print(df3[df3['order'] == 8].max(), '\n\n')

print(df3[df3['order'] >= 9].max())
print(df3[df3['order'] >= 9].mean(),'\n\n')

print(df3.mean(), df3.max(), '\n\n')
```
